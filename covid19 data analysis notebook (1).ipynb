### Task 3.3: find maximum infection rate for all of the countries. 

```python
Countries=list(df1.index)
max_infection=[]
for i in Countries:
    max_infection.append(df1.loc[i].diff().max())
print(max_infection)
```

```python
df1["Max_Infection"]=max_infection
df1.head()
```

### Task 3.4: create a new dataframe with only needed column

### Task 5.1: Plotting GDP vs maximum Infection rate

```python
x =data["GDP per capita"]
y= data["Max_Infection"]
sns.scatterplot(x,y)
```

As we can see that, there is an scaling issuein the above mentioned scatter plot to solve this issue this problem we use **log scaling**

```python
sns.scatterplot(x,np.log(y))
```

```python
sns.regplot(x,np.log(y))
```

We have found a very interesting insight from this analysis that is people living developed countries are prone to getting the infection compared to less developed countries, this might be because of the less testing in less developed countires.
